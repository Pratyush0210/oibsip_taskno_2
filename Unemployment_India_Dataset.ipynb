## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px  # install plotly library and Download the ipynb file and run this code in ur notebook to get the interactive chart. 
%matplotlib inline
```

## **Loading dataset**

```python
unemp_data = pd.read_csv("/content/drive/MyDrive/OASIS INFOBYTE/TASK 2/Unemployment in India.csv")
```

```python
unemp_data.head()
```

```python
unemp_data.drop(' Frequency', axis=1, inplace=True)
```

```python
unemp_data.dropna(inplace=True)
```

```python
state_list = list(unemp_data['Region'].unique())
```

```python
state_urban = unemp_data[unemp_data['Area']=='Urban']
state_rural = unemp_data[unemp_data['Area']=='Rural']
```

## **Visualizing Barcharts on the basis of Dates and Est. Employed by Area**

```python
plt.figure(figsize =(20,6))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)

sns.barplot(unemp_data, x = ' Date', y = ' Estimated Employed', hue = 'Area', errorbar= None)
```

## **Visualizing Barcharts on the basis of Dates and Est. Unemployment Rate (%) by Area**

```python
plt.figure(figsize =(20,6))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)

sns.barplot(unemp_data, x = ' Date', y = ' Estimated Unemployment Rate (%)', hue = 'Area', errorbar= None)
```

```python
plt.figure(figsize =(20,6))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)

sns.barplot(unemp_data, x = ' Date', y = ' Estimated Labour Participation Rate (%)', hue = 'Area', errorbar= None)
```

## **Treemap chart**

- It represent data in the **size** of **Estimated Employed**.
- Data is divided into **Country, States, Areas, and Dates**.
- **Color bar** represents the **Estimated Unemployment Rate (%)**.

```python
# Download the ipynb file and run this code in ur notebook to get the interactive chart.

fig = px.treemap(unemp_data, path=[px.Constant("INDIA"), 'Region','Area', ' Date'], values=" Estimated Employed",
                  color=' Estimated Unemployment Rate (%)', hover_data=['Area'],
                  color_continuous_scale = 'icefire',
                  color_continuous_midpoint = np.average(unemp_data[' Estimated Unemployment Rate (%)'], weights=unemp_data[" Estimated Employed"]))
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```
